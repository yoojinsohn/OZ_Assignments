__[문제3]<br>__

다음 리스트에서 최댓값과 최솟값, 그리고 평균을 출력해주세요.
<br>

```
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
```



````python
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]

max_num = max(nums)
min_num = min(nums)
avg_num = sum(nums) / len(nums)

print(f"최댓값: {max_num}")
print(f"최솟값: {min_num}")
print(f"평균값: {avg_num}")
````

__[문제7]__

아래 순서대로 주어진 수강생 이름과 등급을 하나의 딕셔너리로 만드는 코드를 작성해주세요


names : 나이썬, 김자바, 김씨, 박씨샵  
grades :   B       B      A      C


````python
names = ['나이썬', '김자바', '김씨', '박씨샵']
grades = ['B', 'B', 'A', 'C']

# zip 함수로 묶어서 dict로 변환
groups = dict(zip(names, grades))

print(groups)
````
